```python
import pandas as pd
import re
```

```python
keys = pd.read_csv('new_keys.csv')
lgas = pd.read_csv('new_lgas.csv')
data = pd.read_csv('VPA_Open_Space.csv')
```

```python
tester = set()
for lga in list(data['LGA']):
    tester.add(lga)
tester
```

```python
tally_exist = dict()
area_exist = dict()
tally_planned = dict()
area_planned = dict()
tally_pureland = dict()
area_pureland = dict()

for word in list(lgas['LGA names']): 
    tally_exist[word.upper()] = 0 
    area_exist[word.upper()] = 0 
    tally_planned[word.upper()] = 0 
    area_planned[word.upper()] = 0 
    tally_pureland[word.upper()] = 0 
    area_pureland[word.upper()] = 0 

tally_exist
```

```python
reject = ['Cemeteries', 'Natural and semi-natural open space', 'Parks and gardens', 'Recreation corridor', 'Services and utilities reserves', 'Tertiary institutions']
OS_ACCESS = list(data['OS_ACCESS'])
OS_TYPE = list(data['OS_TYPE'])
OS_CATEGOR = list(data['OS_CATEGOR'])
OS_STATUS = list(data['OS_STATUS'])
SHAPE_Area = list(data['SHAPE_Area'])
LGA = list(data['LGA'])

for i in range(len(LGA)):    
    lga = LGA[i]
    
    if lga == 'MORNINGTON':
        lga = 'MORNINGTON PENINSULA'
    
    if OS_ACCESS[i] != 'Closed' and OS_TYPE[i] != 'Restricted public land' and OS_CATEGOR[i] not in reject:
        
        if OS_STATUS[i] == 'Existing':
            tally_exist[lga] += 1
            area_exist[lga] += SHAPE_Area[i]
    
    if OS_TYPE[i] != 'Restricted public land' and OS_CATEGOR[i] not in reject:
        if OS_STATUS[i] == 'Planned':
            tally_planned[lga] += 1
            area_planned[lga] += SHAPE_Area[i]
    
    if OS_ACCESS[i] != 'Closed' and OS_CATEGOR[i] not in reject:
        
        tally_pureland[lga] += 1
        area_pureland[lga] += SHAPE_Area[i]
        
            
```

```python
LGANAMES = list(lgas['LGA names'])
out = pd.DataFrame({'LGA': LGANAMES, 'Tally exist': list(tally_exist.values()), 
                    'Area exist': list(area_exist.values()), 'Tally planned': list(tally_planned.values()), 
                    'Area planned': list(area_planned.values()), 'Tally pureland': list(tally_pureland.values()), 
                    'Area pureland': list(area_pureland.values())})

out
```

```python
out.to_csv('open_space_tally.csv')
```
