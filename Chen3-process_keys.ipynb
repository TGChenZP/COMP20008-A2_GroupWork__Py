```python
import pandas as pd
import re
```

```python
data = pd.read_csv('lga_draft2.csv')
data1 = pd.read_csv('School.csv')
```

```python
suburbs = set()
for i in range(len(data1[' address_town'])):
    suburbs.add(data1[' address_town'][i].upper())
suburbs = sorted(list(suburbs))
len(suburbs)
suburbs
```

```python
postcode = list()
lga_name = list()
lga_id = list()

for suburb in suburbs:
    for i in range(len(data1['school_type'])):
        if data1[' address_town'][i].upper() == suburb:
            postcode.append(data1[' address_postcode'][i])
            lga_name.append(data1[' lga_name'][i])
            lga_id.append(data1[' lga_id'][i])
            break
            
print(len(postcode))
print(len(lga_name))
print(len(lga_id))
    
```

```python
new_lga_id = list()
for i in lga_id:
    if i == 939:
        new_lga_id.append(29399) #不知道是不是
    elif i<100:
        fixed = int(f"20{str(i)}0")
        new_lga_id.append(fixed)
    else:
        fixed = int(f"2{str(i)}0")
        new_lga_id.append(fixed)
len(new_lga_id)
```

```python
new_names = list()
for name in lga_name:
    name = re.sub(r'\(.\)', '', name).strip()
    name = re.sub(r'\(..\)', '', name).strip()
    new_names.append(name)
    
len(new_names)
```

```python
out = pd.DataFrame({'Postcode': postcode, 'Suburb': suburbs, 'LGA id': new_lga_id, 'LGA names': new_names})
out
```

```python
out.to_csv('lga_postcode.csv')
```
