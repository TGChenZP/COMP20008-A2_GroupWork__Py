```python
tmp = list(data['LGA Name'])[0:80]
LGA = list()
for lga in tmp:
    lga = re.sub(r'-', '', lga)
    lga = re.sub(r'[\(1\)]', '', lga)
    lga = lga.upper()
    LGA.append(lga)
LGA
```
