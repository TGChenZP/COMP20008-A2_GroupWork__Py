```python
import pandas as pd
```

```python
data = pd.read_csv('lgapostcode_draft.csv')
data
```

```python
#just for checking how many lgas there are
set1 = set()
set2 = set()
for i in range(len(list(data['lga_id']))):
    set1.add(list(data['lga_id'])[i])
    set2.add(list(data['lga_name'])[i])

print(len(list(set1)))
print(len(list(set2)))
```

```python
lga_id = list()
lga_name = list()
postal_town = list()
postal_postcode = list()

for i in range(len(list(data['lga_id']))):
    if list(data['lga_name'])[i] not in lga_name:
        lga_id.append(list(data['lga_id'])[i])
        lga_name.append(list(data['lga_name'])[i])
        postal_town.append(list(data['postal_town'])[i].upper())
        postal_postcode.append(list(data['postal_postcode'])[i])



predata = {'lga_id': lga_id, 'lga_name': lga_name, 'postal_town': postal_town, 'postal_postcode': postal_postcode}
predata1 = {'lga_id': lga_id, 'lga_name': lga_name}

out = pd.DataFrame(predata)
#out = pd.DataFrame(predata, columns = columns)
out_a = out.sort_values(ascending=True, by = 'lga_id')
out_a = out_a.set_index('lga_id')
out_b = out.sort_values(ascending=True, by = 'postal_postcode')
out_b = out_b.set_index('postal_postcode')

out_c = pd.DataFrame(predata1)
out_c = out_c.sort_values(ascending=True, by = 'lga_id')
out_c
```
