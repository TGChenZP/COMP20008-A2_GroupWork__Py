```python
import pandas as pd
import re
```

```python
data = pd.read_csv("lga_draft2.csv")
data
```

```python
lga_upper = list()
for lga in list(data['lga_name']):
    lga = re.sub(r'-', ' ', lga).strip()
    lga_upper.append(lga.upper())
```

```python
out = pd.DataFrame({"LGA id":data['lga_id'] ,"LGA names": lga_upper})
out
```

```python
out.to_csv("new_lgas.csv")
```
