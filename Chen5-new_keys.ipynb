```python
import pandas as pd
import re
```

```python
data = pd.read_csv("lga_postcode.csv")
data
```

```python
lga_upper = list()
for lga in list(data['LGA names']):
    lga = re.sub(r'-', ' ', lga).strip()
    lga_upper.append(lga.upper())
```

```python
out = pd.DataFrame({'Postcode':data['Postcode'],"Suburb":data['Suburb'],"LGA id":data['LGA id'] ,"LGA names": lga_upper})
out
```

```python
out.to_csv("new_keys.csv")
```
